```python
record = []

for i in result:
    data = i.text.strip()
    record.append(data)
```

```python
record
```

```python
table=[]

for i in range(len(record)):
    childlist = record[:5]
    del record[:5]
    
    if(len(childlist)==5):
        table.append(childlist)    
```

```python
table[0]
```

```python
table[1]
```

```python
import pandas as pd
```

```python

df = pd.DataFrame(table, columns=['World Rank', 'Country', 'CF', 'Current points', '+/- Rank'])
df.head()
```

```python
df.to_csv("hockey.csv", index=False)
```
